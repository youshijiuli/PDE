求解二维热传导方程

$$\frac{\partial u}{\partial t}-\Delta u=f,   \ (x,y)\in\Omega,0\leq t\leq T$$

边界条件
$$u\big|_{\Gamma_1}=0,
u\big|_{\Gamma_2}=p(y,t)=\frac{\mathrm{e}^{-\pi^2t}\cdot\sin(\pi y)}{\pi^2},u\big|_{\Gamma_3}=-p(y,t),\ 0\leq t \leq T$$


其中$$\Omega\triangleq\left( 0,1\right) \times\left( 0,1\right)$$

$$\Gamma_1\triangleq\left\lbrace \left( x,1\right) \big| 0\leq x\leq1\right\rbrace \cup\left\lbrace \left( x,0\right) \big| 0\leq x\leq1\right\rbrace$$

$$\Gamma_2\triangleq\left\lbrace \left( 0,y\right) \big| 0\leq y\leq1\right\rbrace,
\Gamma_3\triangleq\left\lbrace \left( 1,y\right) \big| 0\leq y\leq1\right\rbrace$$

初始解$$u(x,y,0)=\frac{\cos(\pi x)\cdot\sin(\pi y)}{\pi^2}$$

已知$$f=\mathrm{e}^{-\pi^2t}\cdot\cos(\pi x)\cdot\sin(\pi y)$$

解析解$$u(x,y,t)=\frac{\mathrm{e}^{-\pi^2t}\cdot\cos(\pi x)\cdot\sin(\pi y)}{\pi^2}$$

```python
'''一些必要的模块'''
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
%matplotlib inline
```

```python
'''空间上的均匀剖分'''
box = [0, 1, 0, 1]
N1 = 24
NN = (N1 + 1) ** 2
NC = N1 ** 2 
node = np.zeros((NN, 2))
Y, X = np.mgrid[:box[1]:complex(N1 + 1), :box[3]:complex(N1 + 1)]
node[:, 0] = X.flatten() #按行取出矩阵X中的元素
node[:, 1] = Y.flatten()
cell = np.zeros((2 * NC, 3), dtype = np.int)
sign = np.arange(NN).reshape(N1 + 1, N1 + 1)
cell[:NC, 0] = sign[0:-1, 1:].flatten() #每个小正方形下面的小三角形
cell[:NC, 1] = sign[1:, 1:].flatten()
cell[:NC, 2] = sign[0:-1, 0:-1].flatten()
cell[NC:, 0] = sign[1:, 0:-1].flatten() #每个小正方形上面的小三角形
cell[NC:, 1] = sign[0:-1, 0:-1].flatten()
cell[NC:, 2] = sign[1:, 1:].flatten()
```

```python
'''时间方向的均匀剖分'''
J = [0, 1]
N2 = 4 * N1 ** 2 #均匀剖分的段数
tau = 1 / N2
t = np.linspace(0, 1, N2 + 1)
```

```python
'''生成基函数的梯度'''
v0 = node[cell[:, 2], :] - node[cell[:, 1], :] 
v1 = node[cell[:, 0], :] - node[cell[:, 2], :]
v2 = node[cell[:, 1], :] - node[cell[:, 0], :]
length = np.cross(v2, -v1)
Dlambda = np.zeros((2 * NC, 3, 2), dtype = np.float)
W = np.array([[0, 1], [-1, 0]], dtype = np.int)
Dlambda[:, 0, :] = v0 @ W / length.reshape(-1, 1)
Dlambda[:, 1, :] = v1 @ W / length.reshape(-1, 1)
Dlambda[:, 2, :] = v2 @ W / length.reshape(-1, 1)
```

```python
'''生成总刚度矩阵A'''
A1 = np.zeros((2 * NC, 3, 3))
A = np.zeros((NN, NN))
for i in np.arange(2 * NC):
    for j in np.arange(3):
        for k in np.arange(3):
            A1[i, j, k] = np.dot(Dlambda[i][j], Dlambda[i][k]) * length[i] / 2
for i in np.arange(2 * NC):
    for j in np.arange(3):
        for k in np.arange(3):
            A[cell[i][j], cell[i][k]] = A[cell[i][j], cell[i][k]] + A1[i][j][k]            
```

```python
'''定义一个函数'''
def delta(i, j):
    if i == j:
        delta = 2
    else:
        delta = 1
    return delta
```

```python
'''生成质量矩阵M'''
M1 = np.zeros((2 * NC, 3, 3))
M = np.zeros((NN, NN))
for i in np.arange(2 * NC):
    for j in np.arange(3):
        for k in np.arange(3):
            M1[i, j, k] = delta(j, k) * length[i] / 24
for i in np.arange(2 * NC):
    for j in np.arange(3):
        for k in np.arange(3):
            M[cell[i][j], cell[i][k]] = M[cell[i][j], cell[i][k]] + M1[i][j][k]
```

```python
'''找出边界节点'''
bound_point_down = np.arange(N1 + 1)
bound_point_left = np.arange((N1 + 1), N1 * (N1 + 1), N1 + 1)
bound_point_right = np.arange((2 * N1 + 1), (N1 + 1) ** 2 - 1, N1 + 1)
bound_point_up = np.arange(N1 * (N1 + 1), (N1 + 1) ** 2)
bound_point_lr = np.append(bound_point_left, bound_point_right)
bound_point = np.concatenate((bound_point_down, bound_point_lr, bound_point_up))
bound_point = np.sort(bound_point)
```

```python
'''生成载荷向量b'''
def f(x,y,t):
    f = np.exp(-(np.pi ** 2) * t) * np.cos(np.pi * x) * np.sin(np.pi * y)
    return f
b = np.zeros((N2 + 1, NN))
B = np.zeros((N2 + 1, 2 * NC, 3))
for i in np.arange(N2 + 1):
    for j in np.arange(2 * NC):
        for k in np.arange(3):
            B[i, j, k] = f(node[cell[:, k], :][j][0], node[cell[:, k], :][j][1], t[i]) * length[j] / 6
for i in np.arange(N2 + 1):
    for j in np.arange(2 * NC):
        for k in np.arange(3):
            b[i, cell[j][k]] = b[i, cell[j][k]] + B[i][j][k]
```

```python
'''算出边界已知点'''
def p(y,t):
    p = np.exp(-(np.pi ** 2) * t) * np.sin(np.pi * y) / (np.pi ** 2)
    return p
b1 = np.zeros((N2 + 1, NN))
for i in np.arange(N2 + 1):
    for j in np.arange(N1 - 1):
        b1[i, bound_point_left[j]] = p(node[bound_point_left[j]][1], t[i])
for i in np.arange(N2 + 1):
    for j in np.arange(N1 - 1):
        b1[i,bound_point_right[j]] = -p(node[bound_point_right[j]][1], t[i])        
```

```python
'''向后欧拉迭代求解数值解U_h'''
A1 = tau * A + M
def u_0(x, y):
    u_0 = np.cos(np.pi * x) * np.sin(np.pi * y) / (np.pi ** 2)
    return u_0

Uh = np.zeros(NN)
for i in np.arange(NN):
    Uh[i] = u_0(node[i][0], node[i][1])
U_h = np.zeros((N2 + 1, NN))
U_h[0] = Uh

for i in np.arange(N2):
    '''更新b2'''
    b2 = M @ U_h[i] + tau * (b[i + 1])
    b2-= np.dot(A1, b1[i + 1])
    for j in np.arange(4 * N1):
        b2[bound_point[j]] = b1[i + 1][bound_point[j]]
    '''更新A1'''
    A1[bound_point, :] = 0
    A1[:, bound_point] = 0
    A1[bound_point, bound_point] = 1
    U_h[i + 1] = np.linalg.solve(A1, b2)
    A1 = tau * A + M
```

```python
'''真解的函数值U'''
def u(x, y, t):
    u = np.exp(-(np.pi ** 2) * t) * np.cos(np.pi * x) * np.sin(np.pi * y) / (np.pi ** 2)
    return u
U = np.zeros((N2 + 1, NN))
for i in np.arange(N2 + 1):
    for j in np.arange(NN):
        U[i][j] = u(node[j][0], node[j][1], t[i])
```

```python
'''真解和数值解的一些固定时刻的三维图'''
fig = plt.figure(figsize = (20, 15))
Y, X = np.mgrid[:box[1]:complex(N1 + 1), :box[3]:complex(N1 + 1)]
u = np.exp(-(np.pi ** 2) * t[0]) * np.cos(np.pi * X) * np.sin(np.pi * Y) / (np.pi ** 2)
ax = fig.add_subplot(4, 2, 1, projection = '3d')
figure = ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap=plt.cm.Blues_r)
ax.set_title("real_solution,t[0]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U")

ax = fig.add_subplot(4, 2, 2, projection = '3d')
figure = ax.plot_surface(X, Y, U_h[0].reshape((N1 + 1), (N1 + 1)), rstride = 1, cstride = 1, cmap = 'rainbow')
ax.set_title("compute_solution,t[0]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U_h")

u = np.exp(-(np.pi ** 2) * t[600]) * np.cos(np.pi * X) * np.sin(np.pi * Y) / (np.pi ** 2)
ax = fig.add_subplot(4, 2, 3, projection = '3d')
figure = ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap=plt.cm.Blues_r)
ax.set_title("real_solution,t[600]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U")

ax = fig.add_subplot(4, 2, 4, projection = '3d')
figure = ax.plot_surface(X, Y, U_h[600].reshape((N1+1),(N1+1)), rstride = 1, cstride = 1, cmap = 'rainbow')
ax.set_title("compute_solution,t[600]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U_h")

u = np.exp(-(np.pi ** 2) * t[1300]) * np.cos(np.pi * X) * np.sin(np.pi * Y) / (np.pi ** 2)
ax = fig.add_subplot(4, 2, 5, projection = '3d')
figure = ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap=plt.cm.Blues_r)
ax.set_title("real_solution,t[1300]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U")

ax = fig.add_subplot(4, 2, 6, projection = '3d')
figure = ax.plot_surface(X, Y, U_h[1300].reshape((N1+1),(N1+1)), rstride = 1, cstride = 1, cmap = 'rainbow')
ax.set_title("compute_solution,t[1300]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U_h")

u = np.exp(-(np.pi ** 2) * t[2000]) * np.cos(np.pi * X) * np.sin(np.pi * Y) / (np.pi ** 2)
ax = fig.add_subplot(4, 2, 7, projection = '3d')
figure = ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap=plt.cm.Blues_r)
ax.set_title("real_solution,t[2000]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U")

ax = fig.add_subplot(4, 2, 8, projection = '3d')
figure = ax.plot_surface(X, Y, U_h[2000].reshape((N1+1),(N1+1)), rstride = 1, cstride = 1, cmap = 'rainbow')
ax.set_title("compute_solution,t[2000]")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("U_h")

plt.show()
```

```python
'''误差的2范数'''
U_1 = np.zeros((N2 + 1, 2 * NC, 3))
for i in np.arange(N2 + 1):
    for j in np.arange(2 * NC):
        U_1[i, j, :] = U_h[i][cell[j, :]]
u_1 = np.zeros((N2 + 1, 2 * NC, 3))        
for i in np.arange(N2 + 1):
    for j in np.arange(2 * NC):        
        u_1[i, j, :] = U[i][cell[j, :]]
'''每一个时间层的误差'''
e_1 = np.zeros(N2 + 1)
u_3 = np.zeros((N2 + 1, 2 * NC, 3))
u_error = (u_1 - U_1) ** 2
for i in np.arange(N2 + 1):
    for j in np.arange(2 * NC):
        u_3[i, j, :] = u_error[i][j] * length[j] / 6
for i in np.arange(N2 + 1):
    e_1[i] = u_3[i].sum()
'''总的时空误差'''
e_2 = np.zeros(N2 + 1)
for i in np.arange(N2):
    e_2[i] = e_2[i] + e_1[i]
    e_2[i + 1] = e_2[i + 1] + e_1[i + 1]
e = np.sum(e_2) * tau / 2
error = np.sqrt(e)
print(error)
```
