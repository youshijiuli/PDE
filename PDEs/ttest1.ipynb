```python
import numpy as np
from fealpy.mesh.simple_mesh_generator import rectangledomainmesh
from fealpy.mesh.TriangleMesh import TriangleMesh,TriangleMeshDataStructure
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
box = [0,1,0,1]
mesh = rectangledomainmesh(box, nx=3, ny=5, meshtype='tri')#三角形剖分
node = mesh.node
cell = mesh.ds.cell
tmesh = TriangleMesh(node,cell)
```

```python
from fealpy.mesh.simple_mesh_generator import squaremesh
N = 1
mesh = squaremesh(0, 1, 0, 1, r=N)
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
mesh.find_node(axes, showindex=True)
mesh.find_edge(axes, showindex=True)
mesh.find_cell(axes, showindex=True)
plt.show()
```

```python
from fealpy.mesh.simple_mesh_generator import unitcircledomainmesh
mesh = unitcircledomainmesh(0.1, meshtype='polygon')
fig = plt.figure()
axes = fig.gca()
mesh.add_plot(axes)
plt.show()
print(mesh.node,mesh.ds.cell)
```

```python
import numpy as np
from meshpy.triangle import MeshInfo, build
from fealpy.mesh.TriangleMesh import TriangleMesh
mesh_info = MeshInfo()
mesh_info.set_points([(0,0), (1,0), (1,1), (0,1)])
mesh_info.set_facets([[0,1], [1,2], [2,3], [3,0]])  
h = 0.05
mesh = build(mesh_info, max_volume=h**2)
point = np.array(mesh.points, dtype=np.float)
cell = np.array(mesh.elements, dtype=np.int)
tmesh = TriangleMesh(point, cell)
fig = plt.figure()
axes = fig.gca()
tmesh.add_plot(axes)
plt.show()
```
